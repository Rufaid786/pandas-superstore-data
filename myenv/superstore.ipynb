# Profit by category

```python
categorywise_profit=df.groupby('Category')['Profit'].sum()
categorywise_profit.plot(kind='bar')
plt.xlabel("category")
plt.ylabel("Profit")
plt.title('Categorywise profit')
plt.show()
```

# Sales by Region

```python
df.columns
```

```python
regionwise_sales=df.groupby('Region')['Sales'].sum()
regionwise_sales.plot(kind='line')
plt.xlabel('Region')
plt.ylabel('Sales')
plt.title('Sales by region')
plt.show()
```

# Total Profit by the day of the week


```python
df.columns
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d-%m-%Y', errors='coerce')
df['Day']=df['Order Date'].dt.day_name()
```

```python
prifit_byday=df.groupby('Day')['Profit'].sum()
prifit_byday.plot(kind='bar')
plt.xlabel('Day')
plt.ylabel('Profit')
plt.title('Total Profit by the day of the week')
plt.show()
```
